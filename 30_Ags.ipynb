```python
# Program untuk menghitung Indeks Massa Tubuh (IMT) atau Body Mass Index (BMI)
# dan menentukan kategori berat badan berdasarkan nilai BMI.
# Program ini meminta input berat badan dalam kilogram dan tinggi badan dalam meter,
# kemudian menghitung BMI dan menampilkan kategori berat badan sesuai dengan standar WHO.

# SRC

# https://navipos.com/begini-cara-menghitung-berat-badan-idela-bagi-pria-dan-wanita/
# https://www.bhinneka.com/blog/cara-menghitung-berat-badan-ideal/

def BMI():
    print("=" * 25)
    print("Hitung IMT / BMI")
    print("=" * 25)

    berat_badan = float(input("Masukkan berat badan (kg): "))
    print("Berat badan (kg):", berat_badan)
    tinggi_badan = float(input("Masukkan tinggi badan (m): "))
    print("Tinggi badan (m):", tinggi_badan)
    jenis_kelamin = (input("Masukkan jenis kelamin (L/P): ")).upper()
    if(jenis_kelamin == "L"):
        print("Jenis kelamin: Laki-laki")
    elif(jenis_kelamin == "P"):
        print("Jenis kelamin: Perempuan")
    else:
        print("Jenis kelamin tidak valid")

    bmi = berat_badan / (tinggi_badan ** 2)
    print("BMI Anda adalah: {:.2f}".format(bmi))

    if bmi < 18.5:
        if bmi < 16:
            if jenis_kelamin == "L":
                print("Kategori: Sangat Kurus (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Sangat Kurus (Perempuan)")
        elif 16 <= bmi < 17:
            if jenis_kelamin == "L":
                print("Kategori: Kurus (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Kurus (Perempuan)")
        else:
            if jenis_kelamin == "L":
                print("Kategori: Sedikit Kurus (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Sedikit Kurus (Perempuan)")
    elif 18.5 <= bmi < 24.9:
        if jenis_kelamin == "L":
            print("Kategori: Normal (Laki-laki)")
        elif jenis_kelamin == "P":
            print("Kategori: Normal (Perempuan)")
    elif 25 <= bmi < 29.9:
        if jenis_kelamin == "L":
            print("Kategori: Gemuk (Laki-laki)")
        elif jenis_kelamin == "P":
            print("Kategori: Gemuk (Perempuan)")
    elif bmi >= 30:
        if bmi >= 40:
            if jenis_kelamin == "L":
                print("Kategori: Obesitas Morbid (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Obesitas Morbid (Perempuan)")
        elif 35 <= bmi < 40:
            if jenis_kelamin == "L":
                print("Kategori: Obesitas Tinggi (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Obesitas Tinggi (Perempuan)")
        elif 30 <= bmi < 35:
            if jenis_kelamin == "L":
                print("Kategori: Obesitas Sedang (Laki-laki)")
            elif jenis_kelamin == "P":
                print("Kategori: Obesitas Sedang (Perempuan)")

BMI()
```
